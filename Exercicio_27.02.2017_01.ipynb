# Expressões Lâmbdas

```python
def size_each(words):
    return [len(w) for w in words]
```

```python
words = ['look', 'so', 'car', 'ice', 'melted']
```

```python
size_each(words)
```

```python
size_each = lambda words: [len(w) for w in words]
```

```python
size_each(words)
```

```python
teste = actfunc=lambda x: x
```

```python
a = 'arroz'
```

```python
teste(a)
```

```python
b = [3,4,5]
```

```python
teste(b)
```
